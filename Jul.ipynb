```julia
using Pkg
using  Graphs
using DelimitedFiles
using GraphsFlows
using ProgressBars
```

```julia
res = zeros(Float64, 1000, 1000)
```

```julia
b = [0 1; 1 0]
```

```julia
a = readdlm("test.txt")

g1 = DiGraph(b)

g = DiGraph(a)


for ed in tqdm(edges(g))
    i = src(ed)
    j = dst(ed)
    f, F = maximum_flow(g, i, j)
    res[i,j] = f
end
```

```julia
writedlm("test2.txt", res)
```
