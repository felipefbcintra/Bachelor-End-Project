```julia
using Pkg
using  Graphs
using DelimitedFiles
using GraphsFlows
using ProgressBars
```

```julia
res = zeros(Float64, 1000, 1000)
```

```julia
a = readdlm("large_no_flow.txt")


g = DiGraph(a)


for ed in tqdm(edges(g))
    i = src(ed)
    j = dst(ed)
    f, F = maximum_flow(g, i, j)
    res[i,j] = f
end
```

```julia
writedlm("large.txt", res)
```
