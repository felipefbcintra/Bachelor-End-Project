```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
from networkx.generators.community import LFR_benchmark_graph
n = 250
tau1 = 3
tau2 = 1.5
mu = 0.5
G = LFR_benchmark_graph(
    n, tau1, tau2, mu, average_degree=5, min_community=20
)
```

```python
# modularity Distribution Medium Communities
l = [nx.community.modularity(G,nx.community.louvain_communities(G)) for _ in range(500)]
```

```python
# x = [c for c in l if  isinstance(c,float)]
# plt.hist(list(range(len(x))), height=x)
l = np.array(l)

fig, ax = plt.subplots()
ax.hist(l,  bins=10)
ax.set_ylabel("modularity", fontsize=12)
ax.set_title("Distribution of  Modularity by the Louvain Algorithm (medium network) n = 500", fontsize=16)
```

```python
((1.96*np.std(l))/10)/np.average(l)
np.mean(l)
```

```python
l = [nx.community.modularity(G,nx.community.louvain_communities(G)) for _ in range(15)]
((1.96*np.std(l))/np.sqrt(15))/np.average(l)

```
